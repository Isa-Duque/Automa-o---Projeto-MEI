```python
import googleapiclient.discovery
import pandas as pd
from datetime import datetime, timedelta
from google.colab import drive

# 1. Configura√ß√µes
API_KEY = "SUA CHAVE"
drive.mount('/content/drive', force_remount=True)

# 2. Lista de nichos que voc√™ quer "escanear"
# Adicione ou remova nichos aqui conforme seu interesse
nichos_para_escanear = ["Beleza", "Sa√∫de", "Curiosidades", "Tecnologia"]

def minerar_oportunidades(nichos):
    youtube = googleapiclient.discovery.build("youtube", "v3", developerKey=API_KEY)
    seis_meses_atras = datetime.now() - timedelta(days=180)
    resultados_finais = []

    for nicho in nichos:
        print(f"Minerando nicho: {nicho}...")
        search_response = youtube.search().list(
            q=nicho,
            part="snippet",
            type="video",
            order="date",  # Mudamos de 'viewCount' para 'date' para pegar v√≠deos novos
            maxResults=50,  # Aumentamos o leque para 50 para ter mais chance de achar canais novos
            publishedAfter=seis_meses_atras.isoformat() + "Z"
        ).execute()


        for item in search_response['items']:
            channel_id = item['snippet']['channelId']
            ch_resp = youtube.channels().list(part="snippet,statistics", id=channel_id).execute()

            if ch_resp['items']:
                canal = ch_resp['items'][0]
                raw_date = canal['snippet']['publishedAt']
                clean_date = raw_date.split('.')[0].replace('Z', '')
                data_criacao = datetime.strptime(clean_date, "%Y-%m-%dT%H:%M:%S")

                # FILTRO: Criado h√° menos de 6 meses E j√° tem algum crescimento
                inscritos = int(canal['statistics'].get('subscriberCount', 0))

                if data_criacao > seis_meses_atras:
                    resultados_finais.append({
                        'Nicho': nicho,
                        'Canal': canal['snippet']['title'],
                        'Data Cria√ß√£o': data_criacao.strftime("%d/%m/%Y"),
                        'Inscritos': inscritos,
                        'Total Views': canal['statistics'].get('viewCount', 0),
                        'Link': f"https://www.youtube.com/channel/{channel_id}"
                    })

    return pd.DataFrame(resultados_finais)

# --- EXECU√á√ÉO ---
df_final = minerar_oportunidades(nichos_para_escanear)

# Adicione isso logo antes de salvar no Excel
df_final = df_final.drop_duplicates(subset=['Link'])

# Salvar no Drive
hoje = datetime.now().strftime("%Y-%m-%d")
caminho = f"/content/drive/My Drive/youtube_oportunidades_{hoje}.xlsx"

if not df_final.empty:
    # Ordenar pelos canais que t√™m MAIS inscritos (os que mais viralizaram)
    df_final = df_final.sort_values(by='Inscritos', ascending=False)
    df_final.to_excel(caminho, index=False)
    print(f"\n‚úÖ Conclu√≠do! Planilha salva com {len(df_final)} canais promissores.")
    display(df_final.head(10))
else:
    print("\n‚ùå Nenhum canal novo encontrado com esses termos.")
```
