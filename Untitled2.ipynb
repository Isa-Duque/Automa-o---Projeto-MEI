```python
import pandas as pd
import zipfile
import os
import urllib.parse
from google.colab import drive

# 1. Configurações de Data para TESTE (Apenas 1 dia para ser rápido)
DATA_TESTE = 20251201

# URL ATUALIZADA (Lote 0 de Dezembro)
url = "https://arquivos.receitafederal.gov.br/dados/cnpj/dados_abertos_cnpj/2025-12/Estabelecimentos0.zip"

# 2. Limpeza e Download Seguro
if os.path.exists('lote_teste.zip'):
    os.remove('lote_teste.zip') # Remove o arquivo que deu erro antes

print("Baixando lote da Receita... Aguarde finalizar.")
# O comando abaixo baixa e garante que não pegamos arquivos vazios
os.system(f"wget --no-check-certificate -q {url} -O lote_teste.zip")

# 3. Montar Drive
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

leads_finais = []

try:
    if not zipfile.is_zipfile('lote_teste.zip'):
        print("❌ ERRO: O arquivo baixado não é um ZIP válido. Verifique o link da Receita.")
    else:
        with zipfile.ZipFile('lote_teste.zip', 'r') as z:
            nome_arquivo = [f for f in z.namelist() if 'ESTABELE' in f.upper()][0]
            with z.open(nome_arquivo) as f:
                # Lendo em blocos
                chunks = pd.read_csv(f, sep=';', encoding='latin-1', header=None,
                                     dtype=str, chunksize=500000)

                for i, chunk in enumerate(chunks):
                    # Filtro: Situação Ativa (02) + Data do Teste + Tem Telefone
                    condicao = (chunk[5] == '02') & \
                               (chunk[6] == str(DATA_TESTE)) & \
                               (chunk[19].notnull())

                    dados = chunk[condicao].copy()
                    if not dados.empty:
                        # 0:CNPJ, 6:Data, 11:CNAE, 18:DDD, 19:FONE
                        selecao = dados[[0, 6, 11, 18, 19]]
                        selecao.columns = ['cnpj', 'data_abertura', 'cnae', 'ddd', 'telefone']
                        leads_finais.append(selecao)
                        print(f"Bloco {i}: leads encontrados...")

        if leads_finais:
            df_final = pd.concat(leads_finais)

            # Criar Link Profissional com Nome (Simulado)
            def preparar_link(row):
                ddd = str(row['ddd']).strip()
                fone = str(row['telefone']).strip()
                msg = "Olá! Vi que abriu sua empresa recentemente. Sucesso! Criei um grupo para ajudar novos MEIs: [LINK_DO_SEU_GRUPO]"
                return f"https://wa.me/55{ddd}{fone}?text={urllib.parse.quote(msg)}"

            df_final['link_whatsapp'] = df_final.apply(preparar_link, axis=1)

            caminho = '/content/drive/MyDrive/TESTE_LEADS_UNICO_DIA.xlsx'
            df_final.to_excel(caminho, index=False)
            print(f"✅ SUCESSO! {len(df_final)} leads de {DATA_TESTE} salvos no Drive.")
        else:
            print("Nenhum lead encontrado para este dia específico.")

except Exception as e:
    print(f"❌ Erro inesperado: {e}")
```
