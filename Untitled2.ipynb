```python
import pandas as pd
import zipfile
import os
import urllib.parse
import random
from google.colab import drive

# --- 1. CONFIGURA√á√ïES ---
DATA_TESTE = "20251201"
LINK_GRUPO = "https://chat.whatsapp.com/SEU_LINK_AQUI"
url = "https://arquivos.receitafederal.gov.br/dados/cnpj/dados_abertos_cnpj/2025-12/Estabelecimentos0.zip"

# --- 2. FUN√á√ïES DE APOIO ---
def gerar_conteudo_variado(row):
    ddd = str(row['ddd1']).strip().replace('.0', '')
    fone = str(row['telefone1']).strip().replace('.0', '')

    s1 = f"Ol√°! Parab√©ns pela abertura da sua empresa. Sou a Isabela e tamb√©m empreendo. Sei que o come√ßo do MEI gera muitas d√∫vidas (nota fiscal, impostos...). Criei uma comunidade gratuita para nos ajudarmos. Te interessa?"
    s2 = f"Oi! Tudo bem? Parab√©ns pelo novo CNPJ! Sou empreendedora e sei que o in√≠cio √© dif√≠cil. Te convido para minha comunidade de apoio, l√° ajudamos com emiss√£o de NF e fornecedores, te interessa?"
    s3 = f"Oi, parab√©ns pelo novo neg√≥cio! üöÄ Vi que voc√™ abriu seu CNPJ agora. Se precisar de ajuda com a parte burocr√°tica ou fornecedores, temos um grupo que se ajuda: {LINK_GRUPO}"

    msg_escolhida = random.choice([s1, s2, s3])
    texto_codificado = urllib.parse.quote(msg_escolhida)
    link_wa = f"https://wa.me/55{ddd}{fone}?text={texto_codificado}"
    return link_wa, msg_escolhida

# --- 3. EXECU√á√ÉO ---
if not os.path.exists('/content/drive'): drive.mount('/content/drive')

if not os.path.exists('lote_final.zip'):
    print("üì• Baixando arquivo da Receita...")
    os.system(f"wget -q {url} -O lote_final.zip")

print("üîç Minerando leads com colunas atualizadas...")
leads = []

# Mapeamento baseado na sua apura√ß√£o
# Colunas: 0:basico, 1:ordem, 2:dv, 4:fantasia, 5:sit, 10:abertura, 18:cep, 19:uf, 20:mun, 21:ddd, 22:tel, 28:email
colunas_alvo = [0, 1, 2, 4, 5, 10, 18, 19, 20, 21, 22, 28]

try:
    with zipfile.ZipFile('lote_final.zip', 'r') as z:
        nome = [f for f in z.namelist() if 'ESTABELE' in f.upper()][0]
        with z.open(nome) as f:
            chunks = pd.read_csv(f, sep=';', encoding='latin-1', header=None,
                                 dtype=str, chunksize=500000, usecols=colunas_alvo)

            for i, chunk in enumerate(chunks):
                # Filtro: Coluna 5 (Situa√ß√£o) == 02 e Coluna 10 (Data In√≠cio) == DATA_TESTE
                condicao = (chunk[5] == '02') & (chunk[10] == DATA_TESTE) & (chunk[22].notnull())
                res = chunk[condicao].copy()

                if not res.empty:
                    # Renomeando conforme sua lista
                    res.columns = [
                        'cnpj_basico', 'cnpj_ordem', 'cnpj_dv', 'nome_fantasia',
                        'situacao', 'data_inicio', 'cep', 'uf', 'municipio',
                        'ddd1', 'telefone1', 'email'
                    ]

                    # Criar CNPJ formatado
                    res['cnpj_completo'] = res['cnpj_basico'] + res['cnpj_ordem'] + res['cnpj_dv']
                    leads.append(res)
                    print(f"Bloco {i}: leads encontrados...")

    if leads:
        df_final = pd.concat(leads)

        print("üîÑ Gerando links de WhatsApp e mensagens...")
        resultados = df_final.apply(gerar_conteudo_variado, axis=1)
        df_final['link_whatsapp'] = [r[0] for r in resultados]
        df_final['mensagem_utilizada'] = [r[1] for r in resultados]

        # Reordenar colunas para ficar organizado
        ordem_colunas = ['cnpj_completo', 'nome_fantasia', 'data_inicio', 'email', 'ddd1', 'telefone1', 'uf', 'municipio', 'cep', 'link_whatsapp', 'mensagem_utilizada']
        df_final = df_final[ordem_colunas]

        # Salvar no Drive
        caminho = f'/content/drive/MyDrive/LEADS_FORMATADOS_{DATA_TESTE}.xlsx'
        df_final.to_excel(caminho, index=False)
        print(f"‚úÖ SUCESSO! {len(df_final)} leads salvos em: {caminho}")
    else:
        print("‚ùå Nenhum lead encontrado para esta data com os novos filtros.")

except Exception as e:
    print(f"‚ùå Erro: {e}")
```

```python
import pandas as pd
import zipfile
import os
import urllib.parse
import random
from google.colab import drive, files

# --- 1. CONFIGURA√á√ïES ---
DATA_TESTE = "20251201"
url = "https://arquivos.receitafederal.gov.br/dados/cnpj/dados_abertos_cnpj/2025-12/Estabelecimentos0.zip"

# --- 2. EXECU√á√ÉO ---
# Garante a montagem do Drive
drive.mount('/content/drive', force_remount=True)

# Download r√°pido se o arquivo n√£o existir localmente
if not os.path.exists('lote_final.zip'):
    print("üì• Baixando arquivo da Receita...")
    os.system(f"wget -q --no-check-certificate {url} -O lote_final.zip")

leads = []
colunas_alvo = [0, 1, 2, 4, 5, 10, 18, 19, 20, 21, 22, 28]

print("üîç Minerando leads (Colunas: Nome Fantasia, E-mail, Munic√≠pio)...")
try:
    with zipfile.ZipFile('lote_final.zip', 'r') as z:
        nome = [f for f in z.namelist() if 'ESTABELE' in f.upper()][0]
        with z.open(nome) as f:
            chunks = pd.read_csv(f, sep=';', encoding='latin-1', header=None,
                                 dtype=str, chunksize=500000, usecols=colunas_alvo)

            for chunk in chunks:
                # Filtro validado: Ativa (02) e Data de In√≠cio (10)
                condicao = (chunk[5] == '02') & (chunk[10] == DATA_TESTE)
                res = chunk[condicao].copy()
                if not res.empty:
                    res.columns = ['cnpj_basico', 'cnpj_ordem', 'cnpj_dv', 'nome_fantasia',
                                   'situacao', 'data_inicio', 'cep', 'uf', 'municipio',
                                   'ddd1', 'telefone1', 'email']
                    leads.append(res)

    if leads:
        df_final = pd.concat(leads, ignore_index=True)
        df_final['cnpj_completo'] = df_final['cnpj_basico'] + df_final['cnpj_ordem'] + df_final['cnpj_dv']

        # Gerar Links
        def criar_link(row):
            ddd = str(row['ddd1']).strip().replace('.0', '').replace('nan', '')
            tel = str(row['telefone1']).strip().replace('.0', '').replace('nan', '')
            if not tel: return ""
            msg = urllib.parse.quote(f"Ol√°! Vi que abriu sua empresa recentemente. Sucesso! Criei um grupo para ajudar novos empreendedores. Te interessa?")
            return f"https://wa.me/55{ddd}{tel}?text={msg}"

        df_final['link_whatsapp'] = df_final.apply(criar_link, axis=1)

        # SALVAMENTO DUPLO
        nome_arquivo = f"LEADS_FINAL_VALIDADO_{DATA_TESTE}.xlsx"

        # 1. Salva no Drive
        caminho_drive = f"/content/drive/MyDrive/{nome_arquivo}"
        df_final.to_excel(caminho_drive, index=False)
        print(f"‚úÖ Salvo no Drive: {caminho_drive}")

        # 2. Oferece para baixar direto no navegador
        print("üì• Preparando download direto...")
        df_final.to_excel(nome_arquivo, index=False)
        files.download(nome_arquivo)

    else:
        print("‚ùå Nenhum lead encontrado para esta data.")
except Exception as e:
    print(f"‚ùå Erro: {e}")
```
