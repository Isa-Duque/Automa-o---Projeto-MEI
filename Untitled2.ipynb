```python
import pandas as pd
import zipfile
import os
import urllib.parse
from google.colab import drive

# 1. Configura√ß√µes de Data para TESTE (Apenas 1 dia para ser r√°pido)
DATA_TESTE = 20251201

# URL ATUALIZADA (Lote 0 de Dezembro)
url = "https://arquivos.receitafederal.gov.br/dados/cnpj/dados_abertos_cnpj/2025-12/Estabelecimentos0.zip"

# 2. Limpeza e Download Seguro
if os.path.exists('lote_teste.zip'):
    os.remove('lote_teste.zip') # Remove o arquivo que deu erro antes

print("Baixando lote da Receita... Aguarde finalizar.")
# O comando abaixo baixa e garante que n√£o pegamos arquivos vazios
os.system(f"wget --no-check-certificate -q {url} -O lote_teste.zip")

# 3. Montar Drive
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

leads_finais = []

try:
    if not zipfile.is_zipfile('lote_teste.zip'):
        print("‚ùå ERRO: O arquivo baixado n√£o √© um ZIP v√°lido. Verifique o link da Receita.")
    else:
        with zipfile.ZipFile('lote_teste.zip', 'r') as z:
            nome_arquivo = [f for f in z.namelist() if 'ESTABELE' in f.upper()][0]
            with z.open(nome_arquivo) as f:
                # Lendo em blocos
                chunks = pd.read_csv(f, sep=';', encoding='latin-1', header=None,
                                     dtype=str, chunksize=500000)

                for i, chunk in enumerate(chunks):
                    # Filtro: Situa√ß√£o Ativa (02) + Data do Teste + Tem Telefone
                    condicao = (chunk[5] == '02') & \
                               (chunk[6] == str(DATA_TESTE)) & \
                               (chunk[19].notnull())

                    dados = chunk[condicao].copy()
                    if not dados.empty:
                        # 0:CNPJ, 6:Data, 11:CNAE, 18:DDD, 19:FONE
                        selecao = dados[[0, 6, 11, 18, 19]]
                        selecao.columns = ['cnpj', 'data_abertura', 'cnae', 'ddd', 'telefone']
                        leads_finais.append(selecao)
                        print(f"Bloco {i}: leads encontrados...")

        if leads_finais:
            df_final = pd.concat(leads_finais)

            # Criar Link Profissional com Nome (Simulado)
            def preparar_link(row):
                ddd = str(row['ddd']).strip()
                fone = str(row['telefone']).strip()
                msg = "Ol√°! Vi que abriu sua empresa recentemente. Sucesso! Criei um grupo para ajudar novos MEIs: [LINK_DO_SEU_GRUPO]"
                return f"https://wa.me/55{ddd}{fone}?text={urllib.parse.quote(msg)}"

            df_final['link_whatsapp'] = df_final.apply(preparar_link, axis=1)

            caminho = '/content/drive/MyDrive/TESTE_LEADS_UNICO_DIA.xlsx'
            df_final.to_excel(caminho, index=False)
            print(f"‚úÖ SUCESSO! {len(df_final)} leads de {DATA_TESTE} salvos no Drive.")
        else:
            print("Nenhum lead encontrado para este dia espec√≠fico.")

except Exception as e:
    print(f"‚ùå Erro inesperado: {e}")
```

```python
import pandas as pd
import zipfile
import os
import urllib.parse
import random
from google.colab import drive

# --- 1. CONFIGURA√á√ïES ---
DATA_TESTE = "20251201" # Ajuste para o dia que desejar
LINK_GRUPO = "https://chat.whatsapp.com/SEU_LINK_AQUI" # COLOQUE SEU LINK AQUI

# URL do lote de Dezembro
url = "https://arquivos.receitafederal.gov.br/dados/cnpj/dados_abertos_cnpj/2025-12/Estabelecimentos0.zip"

# --- 2. FUN√á√ïES DE APOIO ---
def gerar_conteudo_variado(row):
    ddd = str(row['ddd']).strip()
    fone = str(row['telefone']).strip()

    # Suas mensagens ajustadas
    s1 = f"Ol√°! Parab√©ns pela abertura da sua empresa. Sou a Isabela e tamb√©m empreendo. Sei que o come√ßo do MEI gera muitas d√∫vidas (nota fiscal, impostos, fornecedores...). Criei uma comunidade gratuita onde compartilho planilhas e guias pr√°ticos. Se quiser entrar e baixar os materiais: {LINK_GRUPO}"
    s2 = f"Oi! Tudo bem? Parab√©ns pelo novo CNPJ! Sou empreendedora e sei que o in√≠cio √© dif√≠cil. Te convido para minha comunidade de apoio, l√° nos ajudamos gratuitamente com emiss√£o de NF, fornecedores e log√≠stica: {LINK_GRUPO}"
    s3 = f"Oi, parab√©ns pelo novo neg√≥cio! üöÄ Vi que voc√™ abriu seu CNPJ agora. Se precisar de ajuda com a parte burocr√°tica ou fornecedores, temos um grupo de empreendedores que se ajuda: {LINK_GRUPO}"

    msg_escolhida = random.choice([s1, s2, s3])
    texto_codificado = urllib.parse.quote(msg_escolhida)

    link_wa = f"https://wa.me/55{ddd}{fone}?text={texto_codificado}"
    return link_wa, msg_escolhida

# --- 3. EXECU√á√ÉO ---
if not os.path.exists('/content/drive'): drive.mount('/content/drive')

if not os.path.exists('lote_final.zip'):
    print("üì• Baixando arquivo da Receita...")
    os.system(f"wget -q {url} -O lote_final.zip")

print("üîç Minerando leads e gerando links...")
leads = []

try:
    with zipfile.ZipFile('lote_final.zip', 'r') as z:
        nome = [f for f in z.namelist() if 'ESTABELE' in f.upper()][0]
        with z.open(nome) as f:
            # Lendo apenas colunas essenciais para velocidade: 0(cnpj), 5(situacao), 6(data), 18(ddd), 19(fone)
            chunks = pd.read_csv(f, sep=';', encoding='latin-1', header=None,
                                 dtype=str, chunksize=500000, usecols=[0, 5, 6, 18, 19])

            for chunk in chunks:
                condicao = (chunk[5] == '02') & (chunk[6] == DATA_TESTE) & (chunk[19].notnull())
                res = chunk[condicao].copy()
                if not res.empty:
                    res.columns = ['cnpj', 'situacao', 'data', 'ddd', 'telefone']
                    leads.append(res)

    if leads:
        df_final = pd.concat(leads)

        print("üîÑ Aplicando rod√≠zio de mensagens...")
        resultados = df_final.apply(gerar_conteudo_variado, axis=1)
        df_final['link_whatsapp'] = [res[0] for res in resultados]
        df_final['mensagem_utilizada'] = [res[1] for res in resultados]

        # Salvar no Drive
        caminho = f'/content/drive/MyDrive/LEADS_PRONTOS_{DATA_TESTE}.xlsx'
        df_final.to_excel(caminho, index=False)
        print(f"‚úÖ SUCESSO! Planilha gerada com {len(df_final)} leads em: {caminho}")
    else:
        print("‚ùå Nenhum lead encontrado para esta data.")

except Exception as e:
    print(f"‚ùå Erro: {e}")
```
