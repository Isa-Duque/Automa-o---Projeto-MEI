```python
import pandas as pd
import zipfile
import os

url = "https://arquivos.receitafederal.gov.br/dados/cnpj/dados_abertos_cnpj/2025-12/Estabelecimentos0.zip"
print("Baixando Lote 0 para análise definitiva...")
os.system(f"wget -q {url} -O lote0.zip")

try:
    with zipfile.ZipFile('lote0.zip', 'r') as z:
        nome = [f for f in z.namelist() if 'ESTABELE' in f.upper()][0]
        with z.open(nome) as f:
            # Lemos apenas as primeiras 5000 linhas para mapear o arquivo
            df_amostra = pd.read_csv(f, sep=';', encoding='latin-1', header=None, dtype=str, nrows=5000)

            print("\n--- MAPEAMENTO DE COLUNAS ---")
            coluna_data = None
            for col in df_amostra.columns:
                # Procura uma coluna onde os valores comecem com 202 (2024 ou 2025)
                if df_amostra[col].str.startswith('202', na=False).any():
                    coluna_data = col
                    print(f" Achei! A Data de Abertura está na COLUNA: {col}")
                    break

            if coluna_data is not None:
                # Agora que sabemos a coluna real, filtramos o arquivo todo
                print(f"Iniciando extração real usando a coluna {coluna_data}...")

                # Vamos assumir que o telefone está 8 colunas depois da data (padrão da Receita)
                col_ddd = coluna_data + 7
                col_fone = coluna_data + 8
                col_situacao = coluna_data - 1

                # Filtro Final
                filtro = (df_amostra[col_situacao] == '02') & \
                         (df_amostra[coluna_data].str.startswith('202', na=False))

                leads = df_amostra[filtro].iloc[:, [0, col_situacao, coluna_data, col_fone]]
                print(f"Sucesso na amostra! Encontrados {len(leads)} leads.")

                # Salvando o resultado da amostra para você já ver algo
                from google.colab import drive
                drive.mount('/content/drive')
                leads.to_excel('/content/drive/MyDrive/LEADS_DE_VERDADE.xlsx', index=False)
                print("Arquivo 'LEADS_DE_VERDADE.xlsx' salvo no Drive!")
            else:
                print("ERRO: Não encontrei nenhuma coluna com o ano 2024 ou 2025 nesta amostra.")

except Exception as e:
    print(f"Erro técnico: {e}")
finally:
    if os.path.exists('lote0.zip'): os.remove('lote0.zip')
```
