```python
import smtplib
import time
import random
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

def enviar_em_massa(lista_contatos):
    # --- CONFIGURAÇÕES DE ACESSO ---
    host = "smtp.gmail.com"
    port = 587
    usuario = "seu-email@gmail.com"
    senha = "sua-senha-de-app"

    link_whatsapp = "https://wa.me/55XXXXXXXXXXX" # COLOQUE SEU NUMERO AQUI

    try:
        # Inicia a conexão com o servidor uma única vez para ser mais rápido
        server = smtplib.SMTP(host, port)
        server.starttls()
        server.login(usuario, senha)

        for i, contato in enumerate(lista_contatos, 1):
            email_destino = contato['email']
            cnpj_cliente = contato['cnpj']

            # Criando a mensagem
            msg = MIMEMultipart()
            msg['From'] = usuario
            msg['To'] = email_destino
            msg['Subject'] = f"Maquininha e Crédito para seu novo CNPJ ! ({cnpj_cliente})"

            corpo_html = f"""
            <html>
                <body>
                    <p>Olá!</p>
                    <p>Vi que vocês iniciaram as atividades recentemente (CNPJ: {cnpj_cliente}) e passei para desejar <strong>Muito SUCESSO!</strong></p>
                    <p>Sou especialista em serviços financeiros e ajudo novas empresas a garantirem as <strong>melhores taxas de maquininha</strong> do mercado e acesso a <strong>crédito facilitado</strong> para capital de giro.</p>
                    <p>Maquininha <strong>sem ALUGUEL</strong>, Tenho cupom de <strong>82% DE DESCONTO</strong> na aquisição da maquininha e taxa Promocional de 0,53%*</p>
                    <p><a href="{https://wa.me/5519989049006?text=Olá%20Isabela,%20vi%20a%20oferta%20da%20maquininha%20e%20tenho%20interesse!}" style="background-color: #25D366; color: white; padding: 10px 20px; text-decoration: none; border-radius: 5px; font-weight: bold;">Falar com Consultor via WhatsApp</a></p>
                    <p>Grande abraço,<br><strong>Isabela Ferreira</strong><br>SITE: <a href="https://isa-duque.github.io/portal-cnpj-mei/">isa-duque.github.io/portal-cnpj-mei/</a></p>
                </body>
            </html>
            """

            msg.attach(MIMEText(corpo_html, 'html'))

            # Envio
            server.send_message(msg)
            print(f"[{i}/400] Enviado para: {email_destino}")

            # PAUSA DE SEGURANÇA:
            # Espera entre 10 a 20 segundos entre cada e-mail para evitar bloqueio
            tempo_espera = random.randint(10, 20)
            time.sleep(tempo_espera)

        server.quit()
        print("\n--- TODOS OS 400 E-MAILS FORAM PROCESSADOS ---")

    except Exception as e:
        print(f"Erro crítico no sistema: {e}")

# --- COMO ORGANIZAR SEUS CONTATOS ---
# Você pode preencher essa lista manualmente ou carregar de um Excel/CSV
meus_contatos = [
    {"email": "marmitariadodon@gmail.com", "cnpj": "12.345.678/0001-99"},
    {"email": "cliente2@gmail.com", "cnpj": "98.765.432/0001-11"},
    # ... adicione os outros 398 aqui
]

enviar_em_massa(meus_contatos)
```
