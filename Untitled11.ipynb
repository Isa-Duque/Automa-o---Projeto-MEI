```python
import pandas as pd
import zipfile
import os
import urllib.parse
from google.colab import drive, files

# --- 1. CONFIGURA√á√ïES ---
# Data de abertura que voc√™ quer filtrar (formato AAAAMMDD)
DATA_TESTE = "20260101"

# Link oficial est√°vel da Receita Federal
url = "https://arquivos.receitafederal.gov.br/dados/cnpj/dados_abertos_cnpj/2026-01/Estabelecimentos0.zip"

# --- 2. EXECU√á√ÉO ---
drive.mount('/content/drive', force_remount=True)

# Limpa arquivos corrompidos de tentativas anteriores
if os.path.exists('lote_final.zip'):
    os.remove('lote_final.zip')

print("üì• Baixando arquivo da Receita (Link Oficial)...")
# Usando --no-check-certificate para evitar erros de SSL comuns no Colab
os.system(f"wget --no-check-certificate {url} -O lote_final.zip")

if os.path.getsize('lote_final.zip') < 1000:
    print("‚ùå Erro: O download falhou ou o link est√° indispon√≠vel no momento.")
else:
    leads = []
    # 11: CNAE Principal | 27: E-mail
    colunas_alvo = [0, 1, 2, 4, 5, 10, 11, 18, 19, 20, 21, 22, 27]

    print("üîç Minerando leads (Filtro: Ativas + Data + E-mail + CNAE)...")
    try:
        with zipfile.ZipFile('lote_final.zip', 'r') as z:
            nome = [f for f in z.namelist() if 'ESTABELE' in f.upper()][0]
            with z.open(nome) as f:
                # Processamento em blocos para n√£o travar a mem√≥ria
                chunks = pd.read_csv(f, sep=';', encoding='latin-1', header=None,
                                     dtype=str, chunksize=500000, usecols=colunas_alvo)

                for i, chunk in enumerate(chunks):
                    # Filtro: Situa√ß√£o 02 (Ativa) e Data de In√≠cio
                    condicao = (chunk[5] == '02') & (chunk[10] == DATA_TESTE)
                    res = chunk[condicao].copy()

                    if not res.empty:
                        res.columns = [
                            'cnpj_basico', 'cnpj_ordem', 'cnpj_dv', 'nome_fantasia',
                            'situacao', 'data_inicio', 'cnae_principal', 'cep', 'uf',
                            'municipio', 'ddd1', 'telefone1', 'email'
                        ]
                        leads.append(res)
                        print(f"‚úÖ Bloco {i}: leads encontrados...")

        if leads:
            df_final = pd.concat(leads, ignore_index=True)
            df_final['cnpj_completo'] = df_final['cnpj_basico'] + df_final['cnpj_ordem'] + df_final['cnpj_dv']

            # Fun√ß√£o para link do WhatsApp
            def criar_link(row):
                ddd = str(row['ddd1']).strip().replace('.0', '').replace('nan', '')
                tel = str(row['telefone1']).strip().replace('.0', '').replace('nan', '')
                if not tel or tel == '': return ""
                msg = urllib.parse.quote(f"Ol√°! Vi que abriu sua empresa no ramo {row['cnae_principal']}. Sucesso! Gostaria de participar de um grupo de MEIs?")
                return f"https://wa.me/55{ddd}{tel}?text={msg}"

            df_final['link_whatsapp'] = df_final.apply(criar_link, axis=1)

            # Reorganizar para o Excel
            colunas_ordenadas = [
                'cnpj_completo', 'nome_fantasia', 'cnae_principal', 'email',
                'ddd1', 'telefone1', 'link_whatsapp', 'uf', 'municipio'
            ]
            df_final = df_final[colunas_ordenadas]

            # Salvar
            nome_arq = f"LEADS_EXTRACAO_{DATA_TESTE}.xlsx"
            df_final.to_excel(nome_arq, index=False)
            df_final.to_excel(f"/content/drive/MyDrive/{nome_arq}", index=False)

            print(f"üöÄ Sucesso! {len(df_final)} leads salvos.")
            files.download(nome_arq)
        else:
            print("‚ùå Nenhum lead encontrado para os crit√©rios definidos.")

    except Exception as e:
        print(f"‚ùå Erro ao processar o arquivo: {e}")
```
